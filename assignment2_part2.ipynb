Denote:

* r = We will go swimming

Express this knowledge base using propositional logic using the expression syntax used in logic.ipynb. You can incoprorate any code you need from logic.ipynb and logic.py.
````python
kb = PropKB()

p, q, r, s, t = expr('p, q, r, s, t')

kb.tell(~p)
kb.tell(q)
kb.tell(expr('p <=> r'))
kb.tell(expr('~r ==> s'))
kb.tell(expr('s ==> t'))

qry = expr('~p ==> t')
print(pl_resolution(kb, qry))
````
